```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"D:\Learnerea\Tables\cancer_data_cleaned.xlsx")
df.head()
```

```python
X = df.drop('diagnosis',axis=1)
y = df.diagnosis
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
elasticReg = ElasticNet(alpha=0.1,l1_ratio=0.5,random_state=42)
```

```python
elasticReg.fit(X_train,y_train)
```

```python
elasticReg.coef_
```

```python
featureImp = [(feature,coef) for feature, coef in zip(X.columns,elasticReg.coef_)]
impDF = pd.DataFrame(featureImp,columns=['features','importance'])
impDF.sort_values(by='importance',ascending=False)
```
