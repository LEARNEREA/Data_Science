```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge 
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"D:\Learnerea\Tables\cancer_data_cleaned.xlsx")
df.head()
```

```python
X = df.drop('diagnosis',axis=1)
y = df.diagnosis
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.33, random_state=42)
```

```python
ridge_reg = Ridge(alpha=1.0)
```

```python
ridge_reg.fit(X_train,y_train)
```

```python
ridge_reg.coef_
```

```python
feature_importance = [(feature,coef) for feature,coef in zip(X.columns,ridge_reg.coef_)]
feature_importance
```

```python
fi_df = pd.DataFrame(feature_importance,columns=['features','coef'])
fi_df.sort_values(by='coef',ascending=False)
```
