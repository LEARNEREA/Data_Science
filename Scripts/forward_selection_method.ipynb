```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
df = pd.read_csv(r"D:\Learnerea\Tables\cancer_data.csv")
df.head()
```

```python
df_new = df.drop('id',axis=1)
df_new.head()
```

```python
df_new['diagnosis'] = np.where(df_new['diagnosis']=='M',1,0)
```

```python
df_new.to_excel(r"D:\Learnerea\Tables\cancer_data_cleaned.xlsx",index=False)
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_new.drop('diagnosis',axis=1), df_new.diagnosis, test_size=0.20, random_state=42)
```

```python
y_train.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
feature_selector = SequentialFeatureSelector(LogisticRegression(),
                                            k_features=(5,20),
                                            forward=True,
                                            floating=False,
                                            verbose=4,
                                            scoring="accuracy",
                                            cv=5).fit(X_train,y_train)
```
