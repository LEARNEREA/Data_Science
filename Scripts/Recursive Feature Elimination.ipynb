```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import RFE, RFECV

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel(r"D:\Learnerea\Tables\cancer_data_cleaned.xlsx")
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('diagnosis',axis=1), df.diagnosis, test_size=0.33, random_state=42)
```

```python
rfe_select = RFE(DecisionTreeClassifier(),n_features_to_select=0.25).fit(X_train,y_train)
```

```python
rfe_select.support_
```

```python
important_features = pd.DataFrame(list(zip(X_train.columns,rfe_select.support_)),columns=['features','important'])
```

```python
important_features.sort_values(by='important',ascending=False)
```
